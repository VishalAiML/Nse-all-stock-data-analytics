```python
# Drop columns with 100% missing data
data_cleaned = data.dropna(axis=1, how='all')

# Check the extent of missing data again
missing_data_cleaned = data_cleaned.isnull().sum().sort_values(ascending=False)
missing_data_percentage_cleaned = (missing_data_cleaned / len(data_cleaned)) * 100

# Display columns with the highest percentage of missing data
missing_data_percentage_cleaned.head(20)
```

```python
threshold = 0.95
data_cleaned = data_cleaned.loc[:, data_cleaned.isnull().mean() < threshold]
```

```python
missing_data_cleaned = data_cleaned.isnull().sum().sort_values(ascending=False)
missing_data_percentage_cleaned = (missing_data_cleaned / len(data_cleaned)) * 100
print(missing_data_percentage_cleaned.head(20))
```

```python
#Relative Strength Index)

def calculate_rsi(data, window):
    delta = data.diff()
    gain = (delta.where(delta > 0, 0)).fillna(0)
    loss = (-delta.where(delta < 0, 0)).fillna(0)
    avg_gain = gain.rolling(window=window, min_periods=1).mean()
    avg_loss = loss.rolling(window=window, min_periods=1).mean()
    rs = avg_gain / avg_loss
    rsi = 100 - (100 / (1 + rs))
    return rsi

data_cleaned['RELIANCE_RSI'] = calculate_rsi(data_cleaned['RELIANCE'], 14)

# Plot RSI
plt.figure(figsize=(14, 7))
plt.plot(data_cleaned['RELIANCE_RSI'], label='RSI')
plt.axhline(70, color='r', linestyle='--')
plt.axhline(30, color='g', linestyle='--')
plt.xlabel('Date')
plt.ylabel('RSI')
plt.title('RELIANCE RSI (14-day)')
plt.legend()
plt.show()
```

```python
#Relative Strength Index)

def calculate_rsi(data, window):
    delta = data.diff()
    gain = (delta.where(delta > 0, 0)).fillna(0)
    loss = (-delta.where(delta < 0, 0)).fillna(0)
    avg_gain = gain.rolling(window=window, min_periods=1).mean()
    avg_loss = loss.rolling(window=window, min_periods=1).mean()
    rs = avg_gain / avg_loss
    rsi = 100 - (100 / (1 + rs))
    return rsi

data_cleaned['RELIANCE_RSI'] = calculate_rsi(data_cleaned['RELIANCE'], 30)

# Plot RSI
plt.figure(figsize=(14, 7))
plt.plot(data_cleaned['RELIANCE_RSI'], label='RSI')
plt.axhline(70, color='r', linestyle='--')
plt.axhline(30, color='g', linestyle='--')
plt.xlabel('Date')
plt.ylabel('RSI')
plt.title('RELIANCE RSI (30-day)')
plt.legend()
plt.show()
```

```python
# Filter data for the year 2023
data_2023 = data_cleaned.loc['2023-01-01':'2023-12-31']

# Calculate the percentage change for each stock
percentage_change = (data_2023.iloc[-1] - data_2023.iloc[0]) / data_2023.iloc[0] * 100

# Find the stock with the highest percentage change
best_stock = percentage_change.idxmax()
best_stock_growth = percentage_change.max()

print(f"The best performing stock in 2023 is {best_stock} with a growth of {best_stock_growth:.2f}%")

# Display the percentage change of all stocks
print("\nPercentage change of all stocks in 2023:")
print(percentage_change.sort_values(ascending=False))
```

```python
# Find the top 10 stocks with the highest percentage change
top_10_stocks = percentage_change.nlargest(10)

top_10_stocks  
```

```python
# Find the stock with the lowest (worst) percentage change
worst_stock = percentage_change.idxmin()
worst_stock_growth = percentage_change.min()

# Find the bottom 10 stocks with the worst percentage change
bottom_10_stocks = percentage_change.nsmallest(10)

print(f"The worst performing stock in 2023 is {worst_stock} with a decline of {worst_stock_growth:.2f}%")

# Print the bottom 10 stocks
print("\nBottom 10 companies with the worst performance in 2023:")
print(bottom_10_stocks)
```
